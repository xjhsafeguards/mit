```python
pdos_cl = np.loadtxt("cl.pdos_atm#1(Cl)_wfc#2(p)")
pdos_cls = np.loadtxt("cl.pdos_atm#1(Cl)_wfc#1(s)")
pdos_cld = np.loadtxt("cl.pdos_atm#1(Cl)_wfc#3(d)")
dos = np.loadtxt("cl.pdos_tot")
```

```python
pdos_cl_c = np.histogram([],500,(-30,0),weights=[])[0]
pdos_cls_c = np.histogram([],500,(-30,0),weights=[])[0]
pdos_cld_c = np.histogram([],500,(-30,0),weights=[])[0]
dos_c,edges = np.histogram([],500,(-30,0),weights=[])

files = !ls | grep "c-"
for f in files:
    pdos_cl = np.loadtxt(f+"/cl.pdos_atm#1(Cl)_wfc#2(p)")
    pdos_cls = np.loadtxt(f+"/cl.pdos_atm#1(Cl)_wfc#1(s)")
    pdos_cld = np.loadtxt(f+"/cl.pdos_atm#1(Cl)_wfc#3(d)")
    dos = np.loadtxt(f+"/cl.pdos_tot")
    pdos_cl_c += np.histogram(pdos_cl[:,0],500,(-30,0),weights=pdos_cl[:,1])[0]
    pdos_cls_c += np.histogram(pdos_cls[:,0],500,(-30,0),weights=pdos_cls[:,1])[0]
    pdos_cld_c += np.histogram(pdos_cld[:,0],500,(-30,0),weights=pdos_cld[:,1])[0]
    dos_c += np.histogram(dos[:,0],500,(-30,0),weights=dos[:,1])[0]
```

```python
bins = 500
ranges = (-30,0)
pdos_cl_q = np.histogram([],bins,ranges,weights=[])[0]
pdos_cls_q = np.histogram([],bins,ranges,weights=[])[0]
pdos_cld_q = np.histogram([],bins,ranges,weights=[])[0]
dos_q,edges = np.histogram([],bins,ranges,weights=[])

files = !ls | grep "q-"
for f in files:
    pdos_cl = np.loadtxt(f+"/cl.pdos_atm#1(Cl)_wfc#2(p)")
    pdos_cls = np.loadtxt(f+"/cl.pdos_atm#1(Cl)_wfc#1(s)")
    pdos_cld = np.loadtxt(f+"/cl.pdos_atm#1(Cl)_wfc#3(d)")
    dos = np.loadtxt(f+"/cl.pdos_tot")
    pdos_cl_q += np.histogram(pdos_cl[:,0],bins,ranges,weights=pdos_cl[:,1])[0]
    pdos_cls_q += np.histogram(pdos_cls[:,0],bins,ranges,weights=pdos_cls[:,1])[0]
    pdos_cld_q += np.histogram(pdos_cld[:,0],bins,ranges,weights=pdos_cld[:,1])[0]
    dos_q += np.histogram(dos[:,0],bins,ranges,weights=dos[:,1])[0]
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2,1,1)
ax.plot(0.5 * (edges[:-1]+edges[1:]),dos_c)
ax.plot(0.5 * (edges[:-1]+edges[1:]),pdos_cl_c)
ax.set_ylim((0,70000))
ax.set_yticks([])
ax.set_xticks(np.arange(-30,0.001,2))
ax = fig.add_subplot(2,1,2)
ax.plot(0.5 * (edges[:-1]+edges[1:]),dos_q)
ax.plot(0.5 * (edges[:-1]+edges[1:]),pdos_cl_q)
ax.set_ylim((0,70000))
ax.set_yticks([])
ax.set_xticks(np.arange(-30,0.001,2))
```
